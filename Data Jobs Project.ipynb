```python
import pandas as pd

url = "https://huggingface.co/datasets/lukebarousse/data_jobs/resolve/main/data_jobs.csv"
df = pd.read_csv(url)

df.head()
```

```python
import ast
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_plot = df['job_title_short'].value_counts().to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_title_short', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Job Title')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
# Only in the Netherlands
df_nl = df[df['job_country'] == 'Netherlands']


df_plot = df_nl['job_title_short'].value_counts().to_frame().reset_index()
df_plot.columns = ['job_title_short', 'count']

sns.set_theme(style='ticks')
sns.barplot(data=df_plot.head(20), x='count', y='job_title_short', hue='count',
            palette='dark:b_r', legend=False)
sns.despine()
plt.title('Top Data Job Titles in the Netherlands')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df['job_country'].value_counts().to_frame().head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_country', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Country')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df['company_name'].value_counts().to_frame()[1:].head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Company')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
# Luke Barousse did the EDA for US, I am going to do an Exploratory Analysis in the Netherlands

# Data Jobs in The Netherlands

df_Data_NL = df[df['job_country'] == 'Netherlands']
df_Data_NL
```

```python
# Data Analyst Jobs in The Netherlands

df_Data_NL = df[(df['job_country'] == 'Netherlands') & (df['job_title_short'] == 'Data Analyst')]
df_Data_NL
```

```python
df_plot = df_Data_NL['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Job Locations for Data Professionals in the Netherlands')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_Data_NL.loc[:, 'location_clean'] = df_Data_NL['job_location'].replace({
    'Netherlands': 'Unspecified',
    'Anywhere': 'Remote',
    'Schiphol': 'Amsterdam Area'
})

df_Data_NL = df[df['job_country'] == 'Netherlands'].copy()

df_plot = df_Data_NL['location_clean'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='location_clean', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Job Locations for Data Professionals in the Netherlands')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df_Data_NL['company_name'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Companies for Data Professionals in the Netherlands')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
